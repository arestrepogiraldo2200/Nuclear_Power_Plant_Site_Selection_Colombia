```python
import geopandas as gpd
import json
import matplotlib.pyplot as plt
```

```python
df = gpd.read_file('./RiesgoIncendios/RiesgoIncendios_CondicionesNormales.shp')
df_alta = df[(df['CLASIFICAC'] == 'Muy Alta')| (df['CLASIFICAC'] == 'Alta') ]#| (df['CLASIFICAC'] == 'Moderada')]
df_alta = df_alta.to_crs(crs=4326)
df_alta = df_alta.reset_index()
df_alta = df_alta.drop(['index'], axis=1)
df_alta
```

```python
df_alta.plot(color = '#be5f004d')
plt.axis('off')
plt.savefig('incendios.png',dpi=500)
```

```python

datajson = []
for i in range(0,len(df_alta)):
    
    area = df_alta["geometry"][i].exterior.coords
    arr = []
    
    for j in range(0,len(area)):
        arr.append(area[j][1])
        arr.append(area[j][0])
        arr.append(0)
        
    datajson.append(arr)
```

```python
save_file3 = open("RiesgoIncendios.json", "w") 
json.dump(datajson, save_file3, indent = 6) 
save_file3.close() 
```
