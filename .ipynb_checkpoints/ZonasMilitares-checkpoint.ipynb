```python
import pandas as pd
import json
import ast
```

```python
df = pd.read_csv('ZonasMilitares/ZONAS_Y_DISTRITOS_MILITARES_EJERCITO_NACIONAL.csv')
df
```

```python
datajson = []
c = df["LOCATION"].apply(ast.literal_eval)

for i in range(0,len(df)):
            
    coords = {'lat': c[i][0], 'lng': c[i][1]}
    datajson.append(coords)
```

```python
save_file3 = open("zonasMilitares.json", "w") 
json.dump(datajson, save_file3, indent = 6) 
save_file3.close() 
```
