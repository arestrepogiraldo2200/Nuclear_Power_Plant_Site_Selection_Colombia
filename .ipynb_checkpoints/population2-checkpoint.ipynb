```python
import pandas as pd
import json
      
df = pd.read_csv("../geonames-all-cities-with-a-population-500.csv", sep=";",  on_bad_lines='skip')
df = df[df['Country'] == "Colombia"]
df = df[['Name', 'Latitude', 'Longitude', 'Population']]
df = df[df['Population']>0]

df = df.reset_index()
df = df.drop(['index'], axis=1)
```

```python
df
```

```python
data = []

for i in range(len(df)):
        
    data.append({ "name": df.iloc[i]['Name'], "latitude" : float(df.iloc[i]['Latitude']), "longitude": float(df.iloc[i]['Longitude']), "pop2023": int(df.iloc[i]['Population'])})
    
save_file3 = open("population.json", "w") 
json.dump(data, save_file3, indent = 6) 
save_file3.close() 
```
