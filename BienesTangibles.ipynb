```python

datajson = []
for i in range(0,len(df)):
    
    area = df["geometry"][i].exterior.coords
    arr = []
    
    for j in range(0,len(area)):
        arr.append(area[j][1])
        arr.append(area[j][0])
        arr.append(0)
        
    datajson.append(arr)
```

```python
save_file3 = open("RiesgoInundacion.json", "w") 
json.dump(datajson, save_file3, indent = 6) 
save_file3.close() 
```
