```python
import pandas as pd
import numpy as np
import json
```

# Versión 1

```python
# Load data
volcanoes = pd.read_csv("../GVP_Volcano_List.csv", delimiter=',')
volcanoes.columns = volcanoes.iloc[0]
volcanoes = volcanoes.drop(0)
volcanoes = volcanoes.reset_index()
volcanoes = volcanoes.drop(['index'], axis=1)
volcanoes = volcanoes[volcanoes['Latitude'].notna()]
volcanoes
```

```python
names = volcanoes['Volcano Name'].to_numpy()
volcanotype = volcanoes['Primary Volcano Type'].to_numpy()
lasteruption = volcanoes['Last Eruption Year'].to_numpy()
lat = volcanoes['Latitude'].to_numpy()
lng = volcanoes['Longitude'].to_numpy()

datajson = []

for i in range(len(names)):
        
    datajson.append({ "name": names[i], "type": volcanotype[i], "last": lasteruption[i], "lat": lat[i], "lng": lng[i]})
    
save_file3 = open("volcanoes.json", "w") 
json.dump(datajson, save_file3, indent = 6) 
save_file3.close() 
```
