```python
import pandas as pd
import numpy as np
import json
```

```python
# Opening JSON file
f = open('LineasLicenciadasElectricidad.json')

# Dictionary
data = json.load(f)  

#print(data[0]["geometry"]["coordinates"])

f.close()
```

```python
print(data[0])
```

```python
def flatten_list(_2d_list):
    flat_list = []
    # Iterate through the outer list
    for element in _2d_list:
        if type(element) is list:
            # If the element is of type list, iterate through the sublist
            for item in element:
                flat_list.append(item)
        else:
            flat_list.append(element)
    return flat_list
    
datajson = []

for i in range(len(data)):
        
    datajson.append(data[i]["geometry"]["coordinates"])
    #print(len(data[i]["geometry"]["coordinates"]))
```

```python
print(datajson[0])
```

```python
print(len(datajson))
```

```python
datajson2 = []

for j in range(0, len(datajson)):
    array = []
                          # Takes 1 out of 2 coordinates in the database
    for i in range(0,len(datajson[j]), 2):
        array.append(datajson[j][i])
    
    datajson2.append(flatten_list(flatten_list(array)))

#print(datajson2)
print(len(datajson2))
```

```python
print(datajson2[0])
```

```python
save_file3 = open("LineasElectricasDepurado.json", "w") 
json.dump(datajson2, save_file3, indent = 6) 
save_file3.close() 
```
