```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
# Opening JSON file
f = open('LineasLicenciadasElectricidad.json')

# Dictionary
data = json.load(f)  

#print(data[0]["geometry"]["coordinates"])

f.close()
```

```python
print(data[0])
```

```python
def flatten_list(_2d_list):
    flat_list = []
    # Iterate through the outer list
    for element in _2d_list:
        if type(element) is list:
            # If the element is of type list, iterate through the sublist
            for item in element:
                flat_list.append(item)
        else:
            flat_list.append(element)
    return flat_list
    
datajson = []

for i in range(len(data)):
        
    datajson.append(data[i]["geometry"]["coordinates"])
    #print(len(data[i]["geometry"]["coordinates"]))
```

```python
print(datajson[0])
```

```python
print(len(datajson))
```

```python
datajson2 = []

for j in range(0, len(datajson)):
    array = []
                          
    for i in range(0,len(datajson[j])):
        array.append(datajson[j][i])
    
    datajson2.append(flatten_list(array))

#print(datajson2)
print(len(datajson2))
```

```python
for j in range(0, len(datajson2)):
    try:
        if type(datajson2[j][0]) is float:
            continue
        else:
            print(datajson2[j])

            datajson2[j] = flatten_list(datajson2[j])
    except:
        print('')
```

```python
print(datajson2[0])
```

```python
datajson3 = []

for j in range(0, len(datajson2)):
    
    line = []
    
    for i in range(0, len(datajson2[j])-3,2):
        
        array = []

        array.append(datajson2[j][i])
        array.append(datajson2[j][i+1])
        array.append(datajson2[j][i+2])
        array.append(datajson2[j][i+3])
        
        line.append(array)
        
    datajson3.append(line)
```

```python
print(datajson3[0])
```

```python
datajson4 = []

for j in range(0, len(datajson3)):
    arr = []
    for i in range(0, len(datajson3[j])):
        if (abs(datajson3[j][i][0]-datajson3[j][i][2]) > 0.2 or abs(datajson3[j][i][1]-datajson3[j][i][3]) > 0.2):
            print('removed jump')
            continue
        arr.append(datajson3[j][i])
    
    datajson4.append(arr) 
```

```python
save_file3 = open("LineasElectricasDepurado.json", "w") 
json.dump(datajson4, save_file3, indent = 6) 
save_file3.close() 
```
