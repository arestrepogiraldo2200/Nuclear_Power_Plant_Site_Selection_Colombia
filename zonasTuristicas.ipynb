```python
with open('zonasTuristicas.txt') as f:
    lines = f.readlines()
```

```python
def extract(test_str):

    sub1 = "("
    sub2 = ")"

    idx1 = test_str.index(sub1)
    idx2 = test_str.index(sub2)
 
    res = ''

    for idx in range(idx1 + len(sub1), idx2):
        res = res + test_str[idx]
        
    return res
```

```python
coords = []

for i in range(len(lines)):
    coords.append(extract(lines[i]).split(','))
    
```

```python
print(coords)
```

```python
for i in range(len(lines)):
    for j in range(2):
        coords[i][j] = float(coords[i][j])
```
